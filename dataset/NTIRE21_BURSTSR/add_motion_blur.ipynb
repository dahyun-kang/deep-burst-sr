```python
def add_random_motionblur(burst):
    # burst: B, C, H, W (1, 4, 80, 80)
    rand_angle = random.uniform(0, 360)
    rand_direction = random.uniform(-1, 1)
    
    blur_kernel = MotionBlur(kernel_size=3, angle=rand_angle, direction=rand_direction)
    return blur_kernel(burst)
```
